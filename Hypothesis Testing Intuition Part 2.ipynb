# DISTRIBUTIONS

### What is a distribution?

A probability distribution is a mathmatical function that, stated in simple terms, can be thought of as providing the probability of occurence of different possible outcomes in an experiment.  

- 'Experiment' for data analysis is picking a random variable from our data. 

In other words, a distribution is a mathmatical function which is linked to the data.  

Tell us what is the likelihood of an event occuring but there is a different approach depending on if the variable is discrete or continuous.    

### Discrete Distributions 

Function that tells us what is the probability that a new observation will be in a certain bucket.

### Continuous Distributions

Characterized by a probability density function PDF. 

Can't find the probability of a specific value because in the world of continuous variables, the probability of one value will always be zero.   

Since the probability of having a specific value is zero (as there are infinite number of possibilities for continuous numbers), correct way to think about it is area under the curve

Can only find probabilities of ranges.

# CENTRAL LIMIT THEOREM
### Populations & Samples

**Population**  
Parameters  
Sigma = Standard deviation of popoulation  

**Sample**  
Subset of population    
Sample is designed to help you assess the parameters; therefore just get statistics  

**Sampling Distribution**  
Probability distribution of sample means

### Central Limit Theorem  
Given that you've taken enough samples (10 is usually enough), the sampling distribution of the sample means will be a normally distributed (regardless of what kind of distribution you had in the original population). 

3 Main Elements:
- sampling distribution of the means is a normal distribution  
- mean of the sampling distribution is equal to the mean of the original population  
- standard deviation of the sampling distribution of the sample mean is linked to standard deviation of the original population (population std deviation / sq root of sample size)

According to the Central Limit Theorem, as the number of observations in the sample grows, the sampling distribution of the sample mean becomes thinner.

# ANALYSIS FRAMEWORK
** Apply Central Limit Theorem**    
Applying Central Limit Theorem we know that...  
- Sampling distribution mean = population mean       
- Sampling distibution std deviation = population standard deviation / SQ   ROOT sample size   

**X Critical**     
Identify critical threshold (given)    
Calculate x critical =  critical threshold / sample size  
We would need to get this amount on average to reach the threshold.    

**Z Score**  
Determine how far x critical is from the sampling distribution mean with the Z-Score  
Z Score = critical threshold - sampling distribution mean / sampling distribution standard deviation  

**P Value**      
Look up Z-score on probability distribution chart to get probability.

### Fed-Ex Case

A client has requested a large freight to be transported urgently from Salt Lake City.  When asked about the weight of the cargo, they could not supply the exact weight, however they have specified that there are a total of 36 boxes.  

The only pane you currently have at Denver is a Cessna 208B Grand Caravan with a max cargo weight of 2,630 lb (1,193 kg).    

Based on this information, what is the probability that all of the cargo can be safely loaded onto the plane and transported?  

This type of cargo follows a mean of 72 lb (32.66kg) and a standard deviation of 3 lb (1.36kg).  

**Given info**  
Population =  all the possible boxes that could come from this client. 
- We know this population has a mean of 72 pounds and   
- standard deviation of 3 pounds    
- Don't know what the underlying distribition is  (most likely not normal)  

They are going to be giving us a sample of 36 boxes  

**Apply Central Limit Theorem**  
Need to look at sampling distribution of the sample means to get probability  
Can't make sampling distribution from scratch but apply the Central Limit Theorem we know that...  

- Sampling distribution Mean = Population Mean =  72
- Sampling distribution Standard Deviation = Population Standard Deviation / sq root sample size = 3/6 = .50  

**X Critical**  
Plane Capacity (Critical Threshold) = 2,640 pounds    
X Critical = 2,640 lb/ 36 boxes = 73.06 lb/box  
The max amount each box can weigh on average to be under the critical threshold.   

** Z Score**  
Need to plot this value somewhere on the sampling distribution  
How far away is X Critical from our Sample Mean?  
What is the probability that the sample they send us is going to have a value beyond x critical?    

Z Score = X Critical - Sampling distribution mean / Sampling Distribution Standard Deviation  
= 73.06 - 72 /0.5 = 2.12 

** Probability Table**  
Look up Z Score 2.12 with table
P(x < x critical) = 0.9830 = 98.3%

Probability that the sample they send us is going to be less than 73.06 pounds is 98.3%. Thus, there is a 98.3% chance the plane can safely take off.

# ONE SAMPLE HYPOTHESIS TESTING

## Population Mean v Sample Mean

Null Hypothesis: population mean, mu = sample mean, mu0 (mu not)    
Alternate Hypotheis = population mean, mu <> sample mean mu0 

### Millenials Case Study
In 2015, millennials were watching 26.5 hours of TV a week with a standard deviation of 10 hours.  Today you surveyed 50 millennials and found that they watch 24 hours of TV per week.  Has the parameter decreased?  

### P-Value Approach

**Step 1: State Null Hypothesis**  
Ho: It has not decreased, i.e. population mean >= 26.5   
H1: It has decreased, i.e. population mean <26.5     

We picked the 50 millennials sample from the population, but the Central Limit Theorem says that ANY or all of the 50 millennial samples we could possibly pick has to come from the sampling distribution of the mean.  

So Applying Central Limit Theorem, we know that...
- Sampling distribution mean = population mean = 26.5  
- Sampling distribution std. deviation = population std. deviation / SQ ROOT sample size = 10 / sq root 50 = 1.41 

** Need to figure out, how likely or extremely unlikely that the observed x bar of 24 was picked as a result of random chance given that the null is correct **

**Step 2: Z Score of X Bar**  
Z Score = x bar - sampling distribution mean / sampling distribution std deviation   
= 24 - 26.5 / 1.41 = -1.77

**Step 3: Determine P Value**  
Looking up -1.77 on probability table, we get 0.0384  
P = 0.0384 = 3.84%


**Step 4: Compare P Value to Level of Significance**  
The probability of picking x bar of 24 is only 3.84% if the null hypothesis is true.

The probability of getting the x bar we observed is so unlikely (less than 5%) that this result could not be just chance - we reject the null hypothesis.  

We reject the null hypothesis with a 95% confidence.   

### Rejection Region Approach  
Alternative way of testing a hypothesis  

**Step 1: State Null Hypothesis**  
Ho: It has not decreased, i.e. population mean >= 26.5   
H1: It has decreased, i.e. population mean <26.5  

How likely or unlikely could the observed x bar of 24 be a result of random chance given that the null is true

** Step 2: Calculate Z Score of X Bar**  
Z Score = x bar - sampling distribution mean / sampling distribution std. deviation  
= 24 - 26.5 = 1.41 = -1.77  

** Step 3: Z Critical**  
Looking up probability of .0495, associated Z Critical is -1.65  
If testing with 5% confidence, can just remember this Z Critical  
Don't have to look up every time because Z Critical doesn't change.  
The likelihood of picking randomly picking a sample from this rejection region is always going to be 5%.    

** Step 4: Compare Z Score to Z Critical **  
Z Score -1.77 < Z Critical -1.65

It is so unlikely that we picked a sample from this rejection region, that something suspicious is going on and that the null hypothesis is not actually correct.

Reject the null hypothesis with 95% confidence.

## Population Proportion vs Sample Proportion
A new version of hypothesis testing  


### American Households Case Study
According to a 2016 survey, a staggering 58% of American households have tablets.  Your manager has asked you test the Hypothesis (with a 97% significance level) that only 58% of American homes have table devices.  You surveyed 100 random households and found that 73 of them had tablets.    

Looking at different sources, the proportion is about 58%.    
Can take any statistic on the internet and can check it is true by surveying your own small random sample.

Want to know if this percentage has grown or not.

**Assumption Check**
- (n)(p hat) >10, households with tablets
- (n)(q hat) > 10,households without tablets

If this assumption is not met, need to increase sample size until it does.   
These assumptions are met so proceed with testing.  

**Step 1: State Null Hypothesis**    

Ho = 58% or fewer households have tablets, i.e. population mean is <= 58%  
H1 = More than 58% of households have tablets, i.e. population mean >58%

population proportion = p = 0.58   
population standard deviation = sq root (p)(q) = sq root (.58)(.42) = .49  

Applying Central Limit Theorem, we know...   
- **mean of the distribution of sample proportions** = population proportion = .58  
- **std deviation of the distribution of sample proportions**  
= population standard  deviation / sq root sample size     

sample proportion (aka p hat) =  73/100 = 0.73  

How likely/unlikely is it that we observed the sample mean proportion of .73 (73/100) by chance given that the null hypothesis is true (that the population proportion is less than or equal to 59%).   

** Step 2: Z Score of P hat**  
Z Score = sample proportion - distribution of sample proportions mean / distribution of sample proportions std deviation  
= 0.73 - 0.58 / 0.049 = 3.06

** Step 3: P Value**  
Looking up Z Score of 3.06, we get p value of 0.9989.

We need area on the right so probability P = 1 - 0.9989 = 0.11%  

There is only a .11% chance of getting our observed sample proportion of 73% - reject the null hypothesis.  

Definitely less than 3% chance (the signficance level we were asked to test).    
Therefore, with 97% confidence,  we reject null hypothesis that 58% or fewer households have tablets.

# STUDENT'S T-DISTRIBUTION (T-Test)  
Deals with situations where your sample size is small.  


### Normal Distribution vs t-distribution
Similar to normal distribution, but t-distribituion has:  

**1. Heavier Tails**
- facilitates outliers in an easier fashion, not as strict as normal distribution 
- tails helps us reject the null hypothesis
- for t-distribution, tail is thicker so 5% will be further down in the tail
- harder to reject the null hypothesis when using t-distribution
- Combats the situation where because your sample size is small, reject the null hypothesis in cases where you shouldn't have.
- As degrees of freedom increases (sample size increases), tails get narrower and converge to normal distribution as df gets to infinity (or around 30).  Df around 30 the t-distribution is pretty much identifcal to normal distribution.

**2. Extra parameter: degrees of freedom = n - 1 **  

**3. Describes a sample**
- larger the sample, the closer to normal distribution. 

### Use the t-distribution (t-test) when:

**BOTH conditions are met**    
1. Population standard deviation is unknown AND
2. The sample size is small (n<30)   

If the standard deviation is unknown, but your sample size is 30 or greater, just use normal distribution. There would be no point in using t-distribution because t-distribution is identical to normal distribution when sample size is 30 or more.  **Can approximate population standard deviation with sample standard deviation.  **  

### Walking Case Study
It is recommended to walk 10,000 steps a day (on average) to be healthy.  You are wondering if on average Americans are meeting this recommendation.  You surveyed 10 random people about how many steps each of them takes per day on average and got the following responses.

We don't know anything about the population here, just assuming mu is 10,000. But what we do know:  
sample mean = x bar = 8,595
sample standard deviation = s = 1,428,36
v = degrees of freedom = 10 -1 = 9  

Normally don't do calcuations about the sample, just proceed to sampling distribution.  But here we have to because we don't know much about the population.

**What you're trying to prove is always alternative hypothesis. **

t critical 0.95, v = 9 = -1.833

### 1 Tail Test
- assuming you have prior knowledge  
- just one to check that one thing is happening on one side 
- easier to prove a one tail test   

### 2 Tail Test   
- Want to see something had an effect that changed the underlying parameter in the popullation.    
- harder to prove a two tail test  
- have to split statistical significance between two tails    
- no prior knowledge, so you have to deal with both sides of the tail

# TWO SAMPLE HYPOTHESIS TESTS

Want to test if there is a difference in two samples from the same population  
Ex: Do left hand and right hand batters show difference in their batting average

## **T-Test**
Common parametric test we might use to compare two sets of data   
Hyptothesis tests, like the t-test, are usually specified in terms of the test-statistic.  
Test statistic reduces your dataset to one number that helps you accept or reject the null hypothesis  


## **Welch's Two Sample t-Test**   
There are a few different versions depending on assumptions: Equal sample size? Equal variance?  

Welch's Two Sample t-Test is the most general as it does not assume either.   
Just assumes the data was sampled at random and that both samples follow a normal distribution.   


**1. State Null  **  
Null Hypothesis: Mean of the two samples are **NOT** equal  
Alternate Hypothesis: Mean of the two samples are equal  

**2. Set P critical ** 

**3. Find P Value   **   
P value: probability of obtaining a test statistic at least as extreme as the one that was actually observed assuming that the null hypothesis is true.  

**4. Compare P value to P Critical **  
**If P < P critical:** Reject null that the mean for the two samples is not equal    
**Else:** Cannot reject the null hypothesis    

**T-Tests in SciPy Stats**    
import scipy.stats  
scipy.stats.test_ind(list_1, list_2, equal_var = False)     
 
Returns tuple(t test statistic , p-value)  

Scipy.stats.ttest_ind assumes a two sided t-test (only tests whether the means of ours samples are different) 

One tail test: one mean is greater than or equal to the other.      
One sided p-value will jsut be half of two sided p-value  

### **Baseball Case Study**

**Hypothesis**
**Null Hypothesis:** left hand and right hand batters show no difference in their batting average  
**Alternate Hypothesis:** left hand and right hand batters show a difference in their batting average

```python
import numpy
import scipy.stats 
import pandas as pd

#Read data into pandas dataframe
baseball_data = pd.read_csv('baseball_stats.csv')  
   
#Split the data set into two data frames - left handed and right handed batters  
baseball_data_left = baseball_data[baseball_data['handedness']=='L']
baseball_data_right= baseball_data[baseball_data['handedness']=='R']  

#Perform Welch's t-test
result = scipy.stats.ttest_ind(baseball_data_left['avg'], baseball_data_right['avg'], equal_var = False)

#Testing decision at 95% confidence
if result[1] <= .05:
   print('REJECT NULL ', result)  
else: 
   print('DO NOT REJECT NULL ', result) 
```

There is some math that says that if we have alarge enough sample size, we can actually use tests that assume normality, even if our data is not normal.  

**Non-Parametric Tests**  
Non-Parametric Test: A statistical test that does not assume our data is drawn from any particular underlying probability distribution.

Tests null hypothesis that two populations are the same   
Tests whether the samples came from the same population, but not which has a higher mean or anything like that. Because of this, it's useful to report these results along with some other information (like the two sample means, medians,etc) 
u,p = scipy.stats.mannwhitneyu(x sample, y sample)
